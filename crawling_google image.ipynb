```python
from selenium import webdriver
from urllib.request import urlopen
from bs4 import BeautifulSoup as bs
from urllib.parse import quote_plus
from selenium.webdriver.common.keys import Keys
import time
import urllib.request
import os
import ssl
```

```python
def createFolder(directory):
    try:
        if not os.path.exists(directory):
            os.makedirs(directory)
    except OSError:
        print ('Error: Creating directory. ' +  directory)
```

```python
keyword = '쓰레기'
createFolder('./'+keyword+'_img_download')  #이미지가 저장될 폴더명 정해줌 "pizza_img_download"

driver = webdriver.Chrome('/Users/parkjuhyeon/chromedriver')
driver.implicitly_wait(3)
```

```python
#구글이미지 검색접속 및 검색어 입력

print(keyword, '검색')
driver.get('https://www.google.co.kr/imghp?hl=ko')

Keyword=driver.find_element_by_xpath('//*[@id="sbtc"]/div/div[2]/input')
Keyword.send_keys(keyword)

driver.find_element_by_xpath('//*[@id="sbtc"]/button').click()
```

```python
#구글에서 스크롤 하는것

elem =  driver.find_element_by_tag_name("body")
for i in range(60): #60초동안 스크롤 내리기 -> 이 숫자 조정해서 크롤링할 이미지 개수 정함
    elem.send_keys(Keys.PAGE_DOWN)  #PAGE_DOWN키로 이미지 더 많이 나올 수 있도록 스크롤 내려줌
    time.sleep(0.5)
    
    #지속적으로 스크롤 하다 보면'결과 더보기' 버튼이 생성되는데, 이 버튼을 눌러주는 코드 추가
try:
    driver.find_element_by_xpath('//*[@id="islmp"]/div/div/div/div[1]/div[4]/div[2]/input').click()
    for i in range(60): #60초동안 스크롤 내리기 -> 이 숫자 조정해서 크롤링할 이미지 개수 정함
        elem.send_keys(Keys.PAGE_DOWN)
        time.sleep(0.5)
except:
    pass
```

```python
links=[]
images = driver.find_elements_by_css_selector("img.rg_i.Q4LuWd")
for image in images:
    if image.get_attribute('src')!=None:
        links.append(image.get_attribute('src'))

print(keyword+' 찾은 이미지 개수:',len(links))
time.sleep(2)
```

```python
# 크롤링한 이미지를 '키워드_number' 형태로 만들어둔 폴더에 저장
# 몇 개 저장되다가 에러나면, 아래 두 코드 실행하고 다시 실행하면 됨
# -> !pip3 install certifi
# -> import ssl
#    ssl._create_default_https_context = ssl._create_unverified_context


for k,i in enumerate(links):
    url = i
    start = time.time()  #코드 실행시간 측정
    urllib.request.urlretrieve(url, "./"+keyword+"_img_download/"+keyword+"_"+str(k)+".jpg")
    print(str(k+1)+'/'+str(len(links))+' '+keyword+' 다운로드 중.. Download time : '+str(time.time() - start)[:5]+' 초')
print(keyword+' --이미지저장 완료--')

driver.close()
```
